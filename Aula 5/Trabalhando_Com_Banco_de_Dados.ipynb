```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:') # Criando o banco de dados local no google collab
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
# quero colocar esses dados dentro do meu banco de dados local
dados.to_sql('clientes', engine, index=False)
```

```python
#inspect serve pra inspecionar o banco de daddos
insp = inspect(engine) # passa o motor do banco de dados
```

```python
print(insp.get_table_names()) # Apareceu a tabela clientes dentro do banco de dados
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"' # queremos todos os que estao empregados
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('Empregados', con=engine, index=False) #Con usado pra especificar ao motor do banco de dados
```

```python
pd.read_sql_table('Empregados', engine)
```

```python
pd.read_sql_table('Empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
url_2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
engine2 = create_engine('sqlite:///:memory:')
```

```python
dados2 = pd.read_csv(url_2)
```

```python
dados2.head()
```

```python
#adcionar no banco de dados
dados2.to_sql('clientes2', engine2, index=False)
```

```python
insp = inspect(engine2)
```

```python
print (insp.get_table_names()) # inspecionando as tabelas presentes no banco de dados
```

```python
pd.read_sql_table('clientes2', engine2)
```

```python
#Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.

query = 'UPDATE clientes2 SET Rendimento_anual = 300000 WHERE ID_Cliente = 6840104'
```

```python
with engine2.connect() as conn:
  conn.execute(text(query))
  conn.commit() # tem que commitar
```

```python
pd.read_sql_table('clientes2', engine2)
```

```python
# Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.

query = 'DELETE FROM clientes2 WHERE ID_Cliente=5008809'
with engine2.connect() as conn:
  result = conn.execute(text(query))
  conn.commit() # tem que commitar
```

```python
pd.read_sql_table('clientes2', engine2)
```

```python
# Criar um novo registro de cliente seguindo as especificações abaixo:
# ID_Cliente: 6850985
# Idade: 33
# Grau_escolaridade: Doutorado
# Estado_civil: Solteiro
# Tamanho_familia: 1
# Categoria_de_renda: Empregado
# Ocupacao: TI
# Anos_empregado: 2
# Rendimento_anual: 290000
# Tem_carro: 0
# Moradia: Casa/apartamento próprio

query = 'INSERT INTO clientes2 (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, Rendimento_anual, Tem_carro, Moradia) VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", 2, 290000, 0, "Casa/apartamento próprio")'
with engine2.connect() as conn:
  conn.execute(text(query))
  conn.commit() # tem que commitar
```

```python
pd.read_sql_table('clientes2', engine2)
```
